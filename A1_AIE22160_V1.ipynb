```python
x = input("Enter a string")
vowels = "aeiou"
vowelc = 0
chars = 0
for c in x:
    if c.isalpha():
        chars+=1
        if c.lower() in vowels:
            vowelc+=1
print(f"Vowels: {vowelc} ")
print(f"consonants: {chars-vowelc}")
```

```python
rows_A = int(input("Enter the number of rows for matrix A: "))
cols_A = int(input("Enter the number of columns for matrix A: "))
matrix_A = []

print(f"Enter the elements for a {rows_A}x{cols_A} matrix:")
for i in range(rows_A):
    row = [float(input(f"Enter element at position ({i+1}, {j+1}): ")) for j in range(cols_A)]
    matrix_A.append(row)

rows_B = int(input("Enter the number of rows for matrix B: "))
cols_B = int(input("Enter the number of columns for matrix B: "))
matrix_B = []

print(f"Enter the elements for a {rows_B}x{cols_B} matrix:")
for i in range(rows_B):
    row = [float(input(f"Enter element at position ({i+1}, {j+1}): ")) for j in range(cols_B)]
    matrix_B.append(row)

result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

for i in range(rows_A):
    for j in range(cols_B):
        for k in range(cols_A):
            result[i][j] += matrix_A[i][k] * matrix_B[k][j]

print("\nMatrix A:")
for row in matrix_A:
    print(row)

print("\nMatrix B:")
for row in matrix_B:
    print(row)

print("\nMatrix AB:")
for row in result:
    print(row)
```

```python
#q3
list1 = input("Enter elements for the first list separated by spaces: ").split()
list2 = input("Enter elements for the second list separated by spaces: ").split()

list1 = [int(element) for element in list1]
list2 = [int(element) for element in list2]

common_elements = set(list1) & set(list2)
common_elements_count = len(common_elements)

print(f"Common Elements Count: {common_elements_count}")
```

```python
rows = int(input("Enter the number of rows for the matrix: "))
cols = int(input("Enter the number of columns for the matrix: "))
original_matrix = []

print(f"Enter the elements for a {rows}x{cols} matrix:")
for i in range(rows):
    row = [float(input(f"Enter element at position ({i+1}, {j+1}): ")) for j in range(cols)]
    original_matrix.append(row)

transposed_matrix = []
for i in range(cols):
    transposed_row = []
    for j in range(rows):
        transposed_row.append(original_matrix[j][i])
    transposed_matrix.append(transposed_row)

print("\nOriginal Matrix:")
for row in original_matrix:
    print(row)

print("\nTransposed Matrix:")
for row in transposed_matrix:
    print(row)
```
